## Imports

```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
```

## Exercício 1


```python
#Criando arrays

#Array 1D
x = np.arange(20)
print 'x:', x

#Array 2D
y = np.arange(30).reshape(5,6)
print 'y: \n', y, '\n'

#Acessando os índices do array
print 'Fatiando duas linhas, permitindo apenas a linha 0, 2, 4: \n', y[0::2], '\n'
print 'Fatiando linhas e colinas, permitindo apenas as 3 primeiras linhas e 4 colunas ao passo de 2: \n', y[:3:, 0:4:2], '\n'

#Acessando os índices do array de maneira invertida
print 'Acessando o penúltimo índice da linha e coluna na matriz y: ', y[-2, -2], '\n'
print 'Invertendo o vetor x, de 19 até 1: \n', x[-1:0:-1], '\n'


#Sclicing
a = y[::2,:]
a[:,:] = 0
print 'a: \n', a, '\n'
print 'y_new: \n', y, '\n'

#Reshape
print 'Tamanho de y:', y.shape
b = y.reshape(6, 5)
print 'Novo tamanho de y:', b.shape, '\n'
print 'Matriz y depois de modificar para 6x5: \n', b
print 'Pode se observar que não é a mesma coisa que a transposição', '\n'

#Transposição
yt = y.T
print 'Transposição do vetor y: \n', yt, '\n'

#Índice de um array
i,j = np.indices( (2, 4) )
print 'i: \n', i
print 'j: \n', j, '\n'

#Operações Matriciais
print 'J multiplicado por 2: \n', j*2, '\n'

g = np.ones((2, 4))
print 'Nova matriz de 2 linhas e 4 colunas de valores iguais a 1: \n', g, '\n'
print 'Soma dos índices das matrizes g e j: \n', g+j, '\n'

h = np.ones((4, 2))*5
print 'Nova matriz de 4 linhas e 2 colunas de valores iguais a 5: \n', h, '\n'
gh = np.dot(g, h)
print 'Multiplicação de 2 matrizes: \n', gh, '\n'
print 'Tamanho da matriz:', gh.shape, '\n'

#Linspace
print 'Linspace de 1 a 10, com passo de 10:', np.linspace(1, 10, 10, endpoint=True)

#Clip
f = np.arange(-5, 6)
print 'vetor f:', f
print 'clip(f,-3,4)', np.clip(f,-3,4)
```

## Exercício 2

```python
w = np.ones((10, 10))
print 'w: \n', w
bl = w[0::2, 0::2]
bl[:,:] = 0
print 'w com adição de zeros a cada 2 linhas e colunas partindo de 0: \n', w, '\n'

p = w[1::2, 1::2]
p[:,:] = 0
print 'w com adição de zeros a cada 2 linhas e colunas partindo de 1: \n', w, '\n'

plt.title('Imagem com matriz 10x10')
plt.imshow(w,cmap='gray')
plt.colorbar()
```

### Visualizando arrays em forma de imagem

```python
w = np.ones((50, 50))
print 'w: \n', w
bl = w[0::2, 0::2]
bl[:,:] = 0
print 'w com adição de zeros a cada 2 linhas e colunas partindo de 0: \n', w, '\n'

p = w[1::2, 1::2]
p[:,:] = 0
print 'w com adição de zeros a cada 2 linhas e colunas partindo de 1: \n', w, '\n'

plt.title('Imagem com matriz 50x50')
plt.imshow(w,cmap='gray')
plt.colorbar()
```

```python
w = np.ones((100, 100))
print 'w: \n', w
bl = w[0::2, 0::2]
bl[:,:] = 0
print 'w com adição de zeros a cada 2 linhas e colunas partindo de 0: \n', w, '\n'

p = w[1::2, 1::2]
p[:,:] = 0
print 'w com adição de zeros a cada 2 linhas e colunas partindo de 1: \n', w, '\n'

plt.title('Imagem com matriz 100x100')
plt.imshow(w,cmap='gray')
plt.colorbar()
```
